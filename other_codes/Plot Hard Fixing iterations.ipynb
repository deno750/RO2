```python
import matplotlib.pyplot as plt
import glob

files = glob.glob("hard_fixing_iterations/*.txt")


for file_path in files:
    f = open(file_path, "r")

    timestamps=[]
    costs=[]
    
    for line in f:
        time,cost=list(map(int,line.split()))
        timestamps.append(time)
        costs.append(cost)
    
    #shift to 0 and convert to second
    start=timestamps[0]
    for i,t in enumerate(timestamps):
        timestamps[i]-=start
        timestamps[i]/=1000
    
    f.close()
    plt.plot(timestamps, costs, marker = 'o')

plt.legend(files)
plt.xlabel("Time")
plt.ylabel("Solution Cost")
plt.savefig('hard_fixing_iterations/plot.png', dpi=600, format='png')
plt.show()
```
